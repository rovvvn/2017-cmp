```python
# import libraries - factorial, matplot
# run this cell to import libraries
import numpy as np
from math import factorial, log, exp
import matplotlib.pyplot as plt
```

```python
# how to change figure size, and font size
# run this cell to adjust figure and font size
plt.rcParams['figure.figsize'] = (20,12)
plt.rcParams.update({'font.size': 15})
```

#1: Na=300, Nb=200, qtotal=100=qA+qB

```python
#Write code as a function of qA

# declare variable expressions, ask user for qA input
qTotal = 100
qA = int(input('Enter amount for quantized energy of atom A (0-100): q(a)= '))
qB = qTotal-qA
Na = 300
Nb = 200

# define omega-A, omega-B, omega-total=(omega-a*omega-b) w/ variables and return each equations
def omegaA(Na,qA):
    return factorial(qA+Na-1)/(factorial(qA)*factorial(Na-1))

def omegaB(Nb,qB):
    return factorial(qB+Nb-1)/(factorial(qB)*factorial(Nb-1))

def omegaTotal(qA):
    return omegaA(Na,qA)*omegaB(Nb,qB)

# display data
print('')
print('q(total):',qTotal)
print('q(a):',qA)
print('q(b):',qB)
print('N(a):',Na,' ','N(b):',Nb)
print('Omega(a):',omegaA(Na,qA))
print('Omega(b):',omegaB(Nb,qB))
print('Omega(total):',omegaTotal(qA))

## plot x-axis=qA, y-axis=omegaTotal
# declare x-axis range, y-axis as an empty list
qA_axis = range(qA+1)
omega_axis = []
    
for x in qA_axis:
    omega_axis.append(omegaTotal(x))

# plot(x,y), label x,y , display title
plt.plot(qA_axis, omega_axis)
plt.title('Total $\Omega$ as $q_{a}$ increases to '+str(qA)+' energy units.')
plt.xlabel('$q_{a}$')
plt.ylabel('$\Omega_{total}$')

plt.show()
```

# PLOT

```python
def calcOmega(N,n):
    return exp(log(factorial(N))-log(factorial(n))-log(factorial(N-n)))
    
N=int(input("Enter N amount of oscillators: "))
N_series = range(N+1)     
Omega = []

for n in N_series:
    Omega.append(calcOmega(N,n))
    
plt.plot(N_series,Omega)
plt.title('N amount of oscillators='+str(N))
plt.xlabel('n')
plt.ylabel('$\Omega(n)$')

plt.show()
```

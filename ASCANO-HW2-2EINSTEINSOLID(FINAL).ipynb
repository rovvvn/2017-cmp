This code will calculate Omega-total from two interacting solids A and B based on atom A's quantized energy units. Also, plot a line graph for Omega-total as a function of qA to see the progression from qA=1 to qA=100 and see how Omega(total) increases and fluctuates.

```python
# import libraries - factorial, matplot, math
# run this cell to import libraries
import numpy as np
from math import factorial, log, exp
import matplotlib.pyplot as plt
```

```python
# how to change figure size, and font size
# run this cell to adjust figure and font size
plt.rcParams['figure.figsize'] = (20,12)
plt.rcParams.update({'font.size': 15})
```

# #1: Na=300, Nb=200, qtotal=100=qA+qB

```python
#Write code as a function of qA

# declare variable expressions, ask user for qA input
qTotal = 100
qA = int(input('Enter amount for quantized energy of atom A (0-100): q(a)= '))
qB = qTotal-qA
Na = 300
Nb = 200

# define omega-A, omega-B, omega-total=(omega-a*omega-b) w/ variables and return each equations
def omegaA(Na,qA):
    return factorial(qA+Na-1)/(factorial(qA)*factorial(Na-1))

def omegaB(Nb,qB):
    return factorial(qB+Nb-1)/(factorial(qB)*factorial(Nb-1))

def omegaTotal(qA):
    return omegaA(Na,qA)*omegaB(Nb,qB)

# display data
print('')
print('q(total):',qTotal)
print('q(a):',qA)
print('q(b):',qB)
print('N(a):',Na,' ','N(b):',Nb)
print('Omega(a):',omegaA(Na,qA))
print('Omega(b):',omegaB(Nb,qB))
print('Omega(total):',omegaTotal(qA))
```

```python
# declare variable expressions, ask for qA input
qTotal = 100
qA = int(input('Enter amount for quantized energy of atom A (0-100): q(a)= '))
qB = qTotal-qA
Na = 3000
Nb = 2000

# define function for omega-A, omega-B, omega-total and return each equations
def omegaA(Na,qA):
    return factorial(qA+Na-1)/(factorial(qA)*factorial(Na-1))

def omegaB(Nb,qB):
    return factorial(qB+Nb-1)/(factorial(qB)*factorial(Nb-1))

def omegaTotal(qA):
    return omegaA(Na,qA)*omegaB(Nb,qB)

# display data
print('')
print('q(total):',qTotal)
print('q(a):',qA)
print('q(b):',qB)
print('N(a):',Na,' ','N(b):',Nb)
print('Omega(a):',omegaA(Na,qA))
print('Omega(b):',omegaB(Nb,qB))
print('Omega(total):',omegaTotal(qA))
```

# #3: PLOT

```python
# Define function 'calcOmega' for another function 'plotOmega' to calculate values for the Omega list #
def calcOmega1(qA):
    
    # declare qB, Na, Nb before the functions(so the functions knows them prior) #
    qB = 100-qA
    Na = 300
    Nb = 200
    
    # Placed both functions that calculates the Omegas into calcOmega.       #
    # The function 'plotOmega' will access these functions later in the code.#
    def approx_omegaA(Na,qA):
        return exp(log(factorial(qA+Na-1))-log(factorial(qA))-log(factorial(Na-1)))

    def approx_omegaB(Nb,qB):
        return exp(log(factorial(qB+Nb-1))-log(factorial(qB))-log(factorial(Nb-1)))
    
    # return Omega(total) to CalcOmega (which goes into the for loop of 'plotOmega'#
    # and stores it) as the first value in the Omega list                          #
    return approx_omegaA(Na,qA)*approx_omegaB(Nb,qB)

#----------------------------------------------------------------------------------------------------#
# this section generates the plot

# define x,y-axes
qA_series = range(qA+1)     
Omega = []

# define function 'plotOmega' to store values into Omega using loop
def plotOmega1(qA):
    qA_series = range(qA+1)     
    Omega = []

    for qA in qA_series:
        Omega.append(calcOmega1(qA))
    
    plt.plot(qA_series,Omega)
    plt.title('$\Omega_{total}$ as $q_{a}$ increases to '+str(qA))
    plt.xlabel('$q_{a}$')
    plt.ylabel('$\Omega_{total}$')

    plt.show
```

```python
plotOmega1(100)
print(calcOmega1(60))
```

# HW 3 - MANDLEBROT SET

This program should produce a graph that represents a fractal (Mandlebrot Set), an infinitely ramified mathematical object that contains a structure within a structure, as deep as we care to see.

z' = z^2 + c

where c = complex contant
      , z = complex number
      , z'= output and put back in equation

```python
# Ronaldo Ascano
# due 9.19.18

# import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import math

# adjust figure, font size
plt.rcParams['figure.figsize'] = (25,20)
plt.rcParams.update({'font.size': 15})

#import urllib.request
#req = urllib.request.Request('http://www-personal.umich.edu/~mejn/cp/data/circular.txt')
#response = urllib.request.urlopen(req)
#the_page = response.read().decode()
#data = np.fromstring(the_page, dtype="float", sep=' ')
#data = np.reshape(data,[500,500])
#mandelbrot_set = np.zeros([i,j], dtype="complex")
#points = 500
#spacing = 4 / points
#mandelbrot_set = np.empty([points,points],dtype=float)


# creates module for this loop
if __name__ == "__main__":
    
    # declare plotting variables and the mandelbrot set as an empty set with NxN dimensions
    delta = 4
    points = 500
    spacing = delta / points
    mandelbrot_set = np.empty([points,points],dtype=np.float32)
    
    for i in range(0, points, 1):
        
        # determines pixelation vertically
        y_space = spacing * (i-points/2)
        
        for j in range(0, points, 1):
            
            # determines pixelation horizontally
            x_space = spacing * (j-points/2)
            
            # start/resets here, start w/ z=0, n counts iterations
            z = 0
            n = 0
            
            # begin while loop, runs continuously
            while True:
                
                # starting with a complex constant we start iteration (at n=0 z=c)
                z = z*z + complex(x_space,y_space)
                
                # adds 1 to n every iteration
                n += 1
                
                # once |z| is greater than or equal to 2 => add n to set
                if abs(z) >= 2:
                    mandelbrot_set[i,j] = n
                    break
                
                # if n reaches a certain # iterations, add 0 to set
                elif n == 180: 
                    mandelbrot_set[i,j] = 0
                    break
                    

# shows the graph, sets the xlim and ylim w/extent, title, label axis
plt.imshow(mandelbrot_set, extent=[-2,2,-2,2], aspect='auto', cmap='hot')
plt.title('The Mandelbrot Set, (a fractal)')
plt.xlabel('x', size=10)
plt.ylabel('y', size=10)
plt.colorbar()
plt.show()
```

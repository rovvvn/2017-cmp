```python
# programmed by: RONALDO ASCANO
# due Wed., Sept. 5, 2018.

# START

# This program will ask the user for an integer N, calculate, and then display the Fibonacci sequence
# starting from N=0 to the specified integer N.

def FibGen(n):
    
    #how to define a function ==> def functionName(variable):
    
    FibList = []
    
    # ^^^ how to declare list ==> name_of_list = []
    
    for x in range(1,n+1):
    
    # range can be specified in for loop
    
        if len(FibList) <= 1:
            
            # "if length of list is less than or equal to 1 then"
            
            FibList.append(1)
            
            # .append - add to list
            
        else:
            
            FibList.append(FibList[-1]+FibList[-2])
            
            # [-1] - access last item
            # [-2] - access second to last item
            # can input 'user-defined' function into append intrinsic function
            
    return FibList

print ('---Fibonacci Series Generator---')

print ('')

N=int(input('Enter an integer for N (0-100): '))

print ('')       

for i in range(0,N+1):
    print('N =',i,':',FibGen(i))
    print('')


# END
```

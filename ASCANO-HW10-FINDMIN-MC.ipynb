```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

x_min, x_max = -2, 2
y_min, y_max = -2, 2
A, B = 1/2, 1/4

f = lambda x, y: A*x**2 + B*y**2

# create mesh grid
nx = np.linspace(x_min,x_max,100)
ny = np.linspace(y_min,y_max,100)
x,y = np.meshgrid(nx,ny)

# 3D
z = f(x, y)
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(x,y,z, cmap='autumn', alpha=0.75)
cset = ax.contour(x,y,z, zdir='z', offset=0)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f')

plt.show()

# split f into its x and y terms as separate functions
def g(x):
    return A*x**2 

def h(y):
    return B*y**2

# apply random x only on the axis and random y only on the y axis
def MonteCarlo(N):
    
    # x-axis
    rand_x = np.random.random(N)
    f1 = np.delete(rand_x,rand_x!=g) # deletes a value not equal(!=) to g
    minimum_x = np.amin(f1) # returns min along x axis.


    # y-axis
    rand_y = np.random.random(N)
    f2 = np.delete(rand_y,rand_y!=h)
    minimum_y = np.amin(f2) # returns min along y axis.
    
    # plugs in min x,y to find min value
    min = f(minimum_x,minimum_y)
    z = min
    
    #print(rand_x[:5], len(rand_x))
    #print(f1[:5], len(f1))
    #print(minimum_x,minimum_y)

    return print('Minimum =', z)
```

```python
MonteCarlo(1000)
```

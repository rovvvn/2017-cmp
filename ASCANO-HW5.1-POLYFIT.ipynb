```python
import matplotlib.pyplot as plt
import numpy as np
from math import exp as exp1

#define a function to base approximation
f = lambda x: x**2 * np.sin(7*x)

#generate plot with random noise using np.random.rand()
a , b = 0 , 2
n = 500
degree = 11

x = np.linspace(a, b, n)
y = f(x) + np.random.rand(n) - 0.5

#fit
fitting = np.polyfit(x,y,degree)
fitting = np.poly1d(fitting)

# resize
plt.figure(figsize=[20, 15])

# plot on same graph
#plt.legend(loc='lower left')
plt.plot(x,y, 'b*', label='noise')
plt.plot(x,f(x), 'r', label='actual')
plt.plot(x, fitting(x), 'g', label='fit')
plt.show()
```
